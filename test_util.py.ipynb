```python
import random

import gym # env
import PIL.Image
import pyvirtualdisplay

from util import create_atari_video

# set virtual display
display = pyvirtualdisplay.Display(visible=0, size=(1400, 900)).start()
```

```python
pacman = gym.make("MsPacman-v0")
```

```python
# show frame
frame = pacman.reset()
PIL.Image.fromarray(frame)
```

```python
# random action policy
class RandomPolicy:
    def __init__(self, action_space):
        self.actions = action_space
        
    def action(self, frame):
        return random.choice(self.actions)
```

```python
# create a policy
policy = RandomPolicy(range(pacman.action_space.n))
policy.action(frame)
```

```python
# test embedded video
create_atari_video("tmp_test_embed_video", pacman, policy, nb_episodes=1)
```
