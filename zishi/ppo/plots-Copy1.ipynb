```python
import matplotlib.pyplot as plt
import seaborn
import numpy as np
import pandas as pd
seaborn.set_style("darkgrid")
```

```python
df1=pd.read_csv("ppo_20210327_200825.csv")
df2 = pd.read_csv("ppo_20210327_203612.csv")
df3 = pd.read_csv("ppo_20210327_214136.csv")
df4 = pd.read_csv("ppo_20210328_021533.csv")
df5 = pd.read_csv("ppo_20210329_022726.csv")
df6= pd.read_csv("ppo_20210405_165311.csv")
ppo_500 = pd.read_csv("ppo_20210405_173212.csv")
ddpg_per=pd.read_csv("ddpg_per_True_20210331_175855.csv")
ddpg_random = pd.read_csv("ddpg_per_False_20210331_155213.csv")
ddpg_500 = pd.read_csv("ddpg_baserun_500stepsmax.csv")
ddpg_new_500 = pd.read_csv("ddpg_per_False_500.csv")
ddpg_300 = pd.read_csv("ddpg_per_False_20210405_183122_300.csv")
```

```python
df6["avg_reward"].plot()
```

```python
plt.figure(figsize= (12,8))
plt.title("PPO Runs with Different Maximum Steps settings \n (100 episodes running avg)",fontsize=15)
plt.style.use("ggplot")
plt.plot(df2["avg_reward"],label="Maximum 300 steps")
plt.plot(ppo_500["avg_reward"],label="Maximum 500 steps")
#plt.plot(df5["avg_reward"][:2000],label="PPO5")
plt.plot(df6["avg_reward"],label="Maximum 1000 steps")
plt.axhline(y=200,color='g',linestyle="--",label="Solved")
plt.xlabel('Episodes', fontsize=15)
plt.ylabel('Rewards', fontsize=15)
plt.legend(loc="lower right", prop={"size":18})
plt.savefig("PPO_steps.png",dpi=400)
plt.show()
```

```python
plt.figure(figsize= (12,8))
plt.title("PPO vs DDPG\n (100 episodes running avg)",fontsize=15)
plt.style.use("ggplot")
plt.plot(df2["avg_reward"],label="PPO")
plt.plot(ddpg_per["avg_reward"],label="DDPG(PER)")
plt.plot(ddpg_random["avg_reward"],label="DDPG(Random)")
plt.axhline(y=200,color='g',linestyle="--",label="Solved")
plt.xlabel('Episodes', fontsize=15)
plt.ylabel('Rewards', fontsize=15)
plt.legend(loc="lower right",prop={"size":18})
plt.savefig("PPO_DDPG.png",dpi=400)
plt.show()
```

```python
plt.figure(figsize= (12,8))
plt.title("DDPG Different Runs\n (100 episodes running avg)",fontsize=15)
plt.style.use("ggplot")
plt.plot(ddpg_300["avg_reward"],label="DDPG maximum 300")
plt.plot(ddpg_new_500["avg_reward"],label="DDPG maximum 500")
plt.plot(ddpg_random["avg_reward"],label="DDPG maximum 1000")
plt.axhline(y=200,color='g',linestyle="--",label="Solved")
plt.xlabel('Episodes', fontsize=15)
plt.ylabel('Rewards', fontsize=15)
plt.legend(loc="lower right")
plt.savefig("DDPG_steps.png",dpi=400)
plt.show()
```
