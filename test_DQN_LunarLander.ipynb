```python
import numpy as np
import PIL.Image
import pyvirtualdisplay

from tfdqn import config
from tfdqn import TFDqnAgent

# set virtual display
display = pyvirtualdisplay.Display(visible=False, size=(1400, 900)).start()
```

```python
# dqn configuration
config
```

```python
# change some settings
config["replay_buffer"]["capacity"] = 10000
config["replay_buffer"]["batch_size"] = 64
config["agent"]["optimizer"]["learning_rate"] = 5e-4
config["agent"]["structure"] = [
    (64, {"activation": "relu"}),
    (64, {"activation": "relu"}),
    (32, {"activation": "relu"}),
]
config["agent"]["gamma"]=0.99
config["agent"]["target"]["soft"] = 1e-3
config["agent"]["target"]["period"] = 4
```

```python
dqn = TFDqnAgent(config)
```

```python
# show env
PIL.Image.fromarray(np.squeeze(dqn.env.render().numpy()))
```

```python
dqn.train(nb_iterations=500, log_interval=10)
```
